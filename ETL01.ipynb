```python
# Funcion para comparar dos columnas del dataset, y devolver un mensaje si son diferentes.
def comparar_columnas(df, col1, col2):    
    if df[col1].equals(df[col2]):
        return f"Las columnas '{col1}' y '{col2}' son iguales."
    else:
        num_diferencias = sum(df[col1] != df[col2])
        return f"Las columnas '{col1}' y '{col2}' son diferentes. Hay {num_diferencias} valores diferentes."
 

def unir_columnas(df, col1, col2, nueva_col):
    # Comprobar que las columnas son iguales
    if not df[col1].equals(df[col2]):
        print(f"Error: las columnas '{col1}' y '{col2}' no son iguales.")
        return df
    # Crear una nueva columna con la concatenación de las dos columnas iguales
    df[nueva_col] = df[col1]
    # Eliminar las columnas originales del DataFrame
    df = df.drop(columns=[col1, col2])
    return df

def unir_columnas_y_reemplazar_nulos(df, col1, col2, nueva_col):
    # Concatenar las dos columnas
    df[nueva_col] = df[col1].fillna(df[col2])
    # Eliminar las columnas originales del DataFrame
    df = df.drop(columns=[col1, col2])
    return df
```

```python
import re
import pandas as pd

# Patrones de nivel a buscar
jr_pattern = r'\b(jr|junior)\b'
mid_pattern = r'\b(mid|middle)\b'
sr_pattern = r'\b(sr|senior)\b'
ssr_pattern = r'\b(ssr|semi\s*senior)\b'

# Función para extraer el nivel de un título de trabajo
def extract_level(title):
    level = ''
    if re.search(jr_pattern, title, flags=re.IGNORECASE):
        level = 'Junior'
    elif re.search(mid_pattern, title, flags=re.IGNORECASE):
        level = 'Mid'
    elif re.search(sr_pattern, title, flags=re.IGNORECASE):
        level = 'Senior'
    elif re.search(ssr_pattern, title, flags=re.IGNORECASE):
        level = 'Semi Senior'
    return level

# Crear una nueva columna 'level' en el DataFrame
df_job_title['level'] = df_job_title['job_title'].apply(extract_level)
df_job_title
```
