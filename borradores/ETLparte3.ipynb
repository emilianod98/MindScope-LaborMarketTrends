```python
def comparar_columnas(df, col1, col2):
    """
    Compara dos columnas de un DataFrame y devuelve un mensaje si son diferentes.
    
    Parameters:
    -----------
    df : DataFrame
        El DataFrame que contiene las columnas a comparar.
    col1 : str
        El nombre de la primera columna a comparar.
    col2 : str
        El nombre de la segunda columna a comparar.
    
    Returns:
    --------
    str
        Si las columnas son iguales, devuelve un mensaje indicando que son iguales.
        Si las columnas son diferentes, devuelve un mensaje indicando que son diferentes
        y la cantidad de valores diferentes.
    """
    
    if df[col1].equals(df[col2]):
        return f"Las columnas '{col1}' y '{col2}' son iguales."
    else:
        num_diferencias = sum(df[col1] != df[col2])
        return f"Las columnas '{col1}' y '{col2}' son diferentes. Hay {num_diferencias} valores diferentes."
```

```python
def unir_columnas(df, col1, col2, nueva_col):
    # Comprobar que las columnas son iguales
    if not df[col1].equals(df[col2]):
        print(f"Error: las columnas '{col1}' y '{col2}' no son iguales.")
        return df
    # Crear una nueva columna con la concatenación de las dos columnas iguales
    df[nueva_col] = df[col1]
        # Eliminar las columnas originales del DataFrame
    df = df.drop(columns=[col1, col2])
    return df
```

```python
df_emi = unir_columnas(df_emi, "gaTrackerData.jobTitle", "header.jobTitle", "job_title" )
```

```python
df_emi = unir_columnas(df_emi, "gaTrackerData.sectorId", "overview.sectorId", "sector_id" )
```

```python
def unir_columnas_y_reemplazar_nulos(df, col1, col2, nueva_col):
    # Concatenar las dos columnas
    df[nueva_col] = df[col1].fillna(df[col2])
    # Eliminar las columnas originales del DataFrame
    df = df.drop(columns=[col1, col2])
    return df
```

```python
df_emi = unir_columnas_y_reemplazar_nulos(df_emi, "gaTrackerData.industry", "overview.industry", "industry" )
```

```python
df_emi = unir_columnas_y_reemplazar_nulos(df_emi, "overview.size", "gaTrackerData.empSize", "industry_size" )
```
